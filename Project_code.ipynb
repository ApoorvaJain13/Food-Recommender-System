# Fuctions to convert title to index and vice versa

```python
def get_index_from_title(title):
    return df[df.title == title]["index"].values[0]
```

```python
def get_title_from_index(index):
    return df[df.index == index]["title"].values[0]
```

# Count Vectorization and Similarity Cosine

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
cv = CountVectorizer()

count_matrix = cv.fit_transform(df['ingredients'])
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
cosine_sim = cosine_similarity(count_matrix)
```

```python
cosine_sim
```

# Sample Data

```python
similar_food_of = "Simon Fischer Fruit Bttr Prune Lekvar"
```

```python
food_index = get_index_from_title(similar_food_of)
```

```python
similar_food = list(enumerate (cosine_sim[food_index]))
```

```python
sorted_similar_food = sorted(similar_food,key = lambda x:x[1], reverse = True)
```

# Results
