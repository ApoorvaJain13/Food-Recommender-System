# Food Item Separation

Separate another column names 'food_item' consisting of only Food Itemsfrom coumns 'brands' and 'title'

```python
df['brand'] = df.brand.astype(str)
```

```python
list_brand = df['brand'].values.tolist()
```

```python
df['title'] = df.title.astype(str)
```

```python
list_title = df['title'].values.tolist()
```

```python
list_fooditem = []

for item1, item2 in zip(list_title,list_brand):
    if item2 in item1:
        item3 = item1.replace(item2,'')
        list_fooditem.append(item3)
```

```python
df['food_item'] = pd.Series(list_fooditem)
df.head()
```

# Fuunctions to get Food Item from index and vice- versa

```python
def get_index_from_food_item(food_item):
    for i in range (0,9083):
        if df.iloc[i]['food_item'] == food_item:
            return df.iloc[i]['index']
```

```python
def get_food_item_from_index(index):
    for i in range (0,9083):
        if df.iloc[i]['index'] == index:
            return df.iloc[i]['food_item']
```

```python
def get_title_from_index(index):
    for i in range (0,9083):
        if df.iloc[i]['index'] == index:
            return df.iloc[i]['title']
```

# Count Vectorization and Similarity Cosine

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
cv = CountVectorizer()

count_matrix = cv.fit_transform(df['ingredients'])
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
cosine_sim = cosine_similarity(count_matrix)
```

```python
cosine_sim
```

# Sample Data

```python
similar_food_of = " Plus Giant Pops"
```

```python
food_index = get_index_from_food_item(similar_food_of)
food_index
```

```python
similar_food = list(enumerate (cosine_sim[food_index]))
```

```python
sorted_similar_food = sorted(similar_food,key = lambda x:x[1], reverse = True)
```
